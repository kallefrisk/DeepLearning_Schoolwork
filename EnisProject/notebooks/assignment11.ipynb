```python
import pandas as pd

def add_frame_overlap_column(full_csv_path, trimmed_csv_path, column_name="running_video"):
    df_full = pd.read_csv(full_csv_path)
    df_trimmed = pd.read_csv(trimmed_csv_path)

    trimmed_frames = set(df_trimmed["FrameNo"].values)

    values = []

    for frame in df_full["FrameNo"]:
        if frame in trimmed_frames:
            values.append(1)
        else:
            values.append(0)

    df_full[column_name] = values


    return df_full
```

```python
import os

# Paths
full_dir = "../../MainProject/Assignment11/data/kinect_good_preprocessed_not_cut "
trimmed_dir = "../../MainProject/Assignment9/data/kinect_good_preprocessed"
output_dir = "../../MainProject/Assignment11/data/kinect_good_preprocessed_not_cut_start_stop"


for file_name in os.listdir(full_dir):
    if file_name.endswith(".csv"):

        full_path = os.path.join(full_dir, file_name)
        trimmed_path = os.path.join(trimmed_dir, file_name)

        if not os.path.exists(trimmed_path):
            print(f"Skipping (no match): {file_name}")
            continue

        df = add_frame_overlap_column(full_path, trimmed_path)

        output_path = os.path.join(output_dir, file_name)
        df.to_csv(output_path, index=False)

        print(f"Saved: {file_name}")
```

```python
import os

# Paths
full_dir = "../../MainProject/Assignment10/data/csv_of_all_videos"
trimmed_dir = "../../MainProject/Assignment9/data/kinect_good_preprocessed"
output_dir = "../../MainProject/Assignment11/data/mediapipe_not_cut_start_stop"



for mp_file in os.listdir(full_dir):

    if mp_file.endswith("_mediapipe.csv"):

        base_id = mp_file.replace("_mediapipe.csv", "")
        kinect_file = f"{base_id}_kinect.csv"
        mp_full_path = os.path.join(full_dir, mp_file)
        kinect_cut_path = os.path.join(trimmed_dir, kinect_file)

        if not os.path.exists(kinect_cut_path):
            print(f"Skipping (no match): {mp_file}")
            continue


        df = add_frame_overlap_column(
            full_csv_path=mp_full_path,
            trimmed_csv_path=kinect_cut_path,

            column_name="running_video"

        )


        output_path = os.path.join(output_dir, mp_file)
        df.to_csv(output_path, index=False)

        print(f"Saved: {mp_file}")
```
